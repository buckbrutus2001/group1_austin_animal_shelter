```python
def change_age(i):
    j = i.split(" ")
    if (j[1] == "months" or j[1] == "month"):
        i = int(j[0])/12
    if (j[1] == 'weeks' or j[1] == 'week'):
        i = int(j[0])/52
    if (j[1] == 'years' or j[1] == 'year'):
        i = int(j[0])
    return i
```

```python
intake_df['age_upon_intake'] = intake_df['age_upon_intake'].apply(change_age)
```

```python
def set_address(i):
    j = i.split(' ')
    try:
        n = int(i[0])
        address = ''
        j.remove('(TX)')
        j.remove('in')
        j.remove('Austin')
        for x in j:
            address = address + x + ' '
        address = address + 'Austin, TX'
    except:
        address = i
    finally:
        return address
        
    
```
